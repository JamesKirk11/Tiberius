```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,w = np.loadtxt("prism_wavelength_solution.dat",unpack=True)
```

```python
width = []
npixels = 512
for i in range(npixels):
    if i == 0 or i == 13:
        width.append((w[i+1]-w[i])/2)
    elif i == 511:
        width.append((w[i]-w[i-1])/2)
    else:
        width.append((w[i+1]-w[i])/2+(w[i]-w[i-1])/2)
width = np.array(width)
```

```python
for i in range(npixels):
    print("Bin centre = %.5f, bin left = %.5f, bin right = %.5f"%(w[i],w[i]-width[i]/2,w[i]+width[i]/2))
```

```python
plt.figure()
plt.errorbar(w,x,xerr=width/2,fmt='.')
plt.show()
```

```python
new_tab = open("prism_wavelength_solution.dat","w")
new_tab.write("# PRISM wavelength solution using ground-based CV3 data by E. Schlawin. Bin widths at 1-pixel resolution added by J. Kirk\n")
new_tab.write("# Pixel, wavelength (um), bin width in 1-pixel-wide bin (um) \n")

for i in range(512):
    new_tab.write("%f %f %f \n"%(x[i],w[i],width[i]))

new_tab.close()
```
